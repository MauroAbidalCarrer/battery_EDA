```python
mean_collect_battery = df["Collect Battery"].mean()
mean_drop_battery = df["Drop Battery"].mean()

print(bat_serials[4])
bat_df = bat_dfs[4]
display(bat_df)

# Calculate charged amount with consideration for missing entries
charged_amounts = []
i = 0
while i < len(bat_df) - 1:
    print("i:", i)
    # Get current and next row's serial numbers to check continuity
    current_collect_serial = bat_df.iloc[i]['Collect Battery Serial']
    next_drop_serial = bat_df.iloc[i + 1]['Drop Battery Serial']
    
    # Check if the current drop leads to the next collect for the same battery
    collect_level = bat_df.iloc[i]['Collect Battery']
    if current_collect_serial == next_drop_serial:
        # Normal case: both collect and drop are present
        drop_level = bat_df.iloc[i + 1]['Drop Battery']
        charged_amount = drop_level - collect_level
        print(f"normal\tcase i: {i}, diff:{charged_amount}, calcul: {drop_level} - {collect_level}")
        i += 1  # Increment to skip the next row as it's part of the current calculation
    else:
        # For the next row, check if it's a collect without a preceding drop in sequence
        charged_amount = mean_drop_battery - collect_level
        print(f"edge\tcase i: {i}, diff:{charged_amount}, calcul: {mean_drop_battery} - {collect_level}")

    charged_amounts.append(charged_amount)
    i += 1
    print()

print(charged_amounts)
print(sum(charged_amounts) // 100)
print(sum(charged_amounts) % 100)
print(sum(charged_amounts) / 100)

print(mean_collect_battery)
print(mean_drop_battery)
```
